```python
import chardet
import pandas as pd

pd.set_option('display.max_columns', None)

with open('/Users/heogeonho/Univ/Datamining/TP/데이터 전처리 코드 및 파일/4_허건호_data/4_허건호02.csv', 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']

data = pd.read_csv('/Users/heogeonho/Univ/Datamining/TP/데이터 전처리 코드 및 파일/4_허건호_data/4_허건호02.csv', encoding=encoding)
df = data

print(df.columns)
```

```python
def map_day_to_period(day):
    if day <= 10:
        return '월초'
    elif day <= 20:
        return '월중순'
    else:
        return '월말'

df['한달내의 시기'] = df['화재발생_일'].apply(map_day_to_period)
```

```python
def map_month_to_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    else:
        return '겨울'

df['계절'] = df['화재발생_월'].apply(map_month_to_season)
df
```

```python
def merge_reasons(row):
    if row['특보내용'] == '건조경보' or row['특보내용'] == '건조주의보' :
        return '특보_건조'
    
    elif row['특보내용'] == '강풍경보' or row['특보내용'] == '강풍주의보' :
        return '특보_강풍'
    elif row['특보내용'] == '풍랑경보' or row['특보내용'] == '풍랑주의보' :
        return '특보_풍랑'
    elif row['특보내용'] == '한파경보' or row['특보내용'] == '한파주의보' :
        return '특보_한파'
    elif row['특보내용'] == '대설주의보' :
        return '특보_대설'
    
df['특보내용'] = df.apply(merge_reasons, axis=1)
```

```python
df = df.drop(['화재발생_월','화재발생_일','현장소방지역대거리','화재발생일시','화재발생_연','읍면동명','소방서명','서센터명','소방지역대명','인명피해수','재산피해금액'],axis=1)

df
```

```python
df = df.reindex(columns=['한달내의 시기', '계절', '주말여부', '화재발생시간대', '시군구명', '발화지점', '날씨', '온도', '습도', '시간단위풍속', '특보내용',
       '발화원인종합', '출동소요시간', '현장소방서거리', '현장안전센터거리', '전체인력수합계', '화재진압시간', '동원장비수',
       '피해면적'])
df
```

```python
df.to_csv('4_허건호03.csv', index=False, encoding='utf-8-sig')
df.to_excel('4_허건호03.xlsx')
```
